### 1.2.1 모든 값이 0인 열 지우기

```python
con_removed = [] # 삭제할 열을 모을 빈 리스트

# 0인 열 삭제 시작
for col in df_train.columns:
    if df_train[col].std() == 0: # std()로 0만 있는 열을 찾음
        con_removed.append(col)

df_train.drop(con_removed, axis=1, inplace=True)
df_test.drop(con_removed, axis=1, inplace=True)

print ('삭제한 열은 {}개'.format(len(con_removed)))
```

### 1.2.2 중복 열 삭제

```python
dup_removed = [] # 삭제 될 중복 열
col_tmp = [] # 열을 검색하기 위해 임시로 보관할 리스트
dupList = {} # 중복 열의 딕셔너리

columns = df_train.columns
for i in range(len(columns)-1):
    v = df_train[columns[i]].values
    dupCols = [] # 중복되는 열을 담기 위한 빈 리스트
    for n in range(i+1, len(columns)): # i와 i+1이 같은지 비교
        if np.array_equal(v, df_train[columns[n]].values):
            dup_removed.append(columns[n])
            if columns[n] not in col_tmp:
                dupCols.append(columns[n])
                dupList[columns[i]] = dupCols
                
df_train.drop(dup_removed, axis=1, inplace=True)
df_test.drop(dup_removed, axis=1, inplace=True)

print ('중복되는 열의 수는 {}개'.format(len(dup_removed)))
```

## 2. 모델 선택

고객 만족을 예측하기 위한 분석 모델을 선택하기 위해 cross_validation을 이용해 성능을 확인
 - 확인할 분류 모델
   - naive_bayes
   - linear_model
   - RandomForestClassifier
   - GradientBoostingClassifier

```python
from sklearn import cross_validation as cv
from sklearn import cross_validation
from sklearn import naive_bayes 
from sklearn import linear_model
from sklearn import ensemble
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
import time
```

### 2.1 분석 성능을 위한 함수 만들기

```python
def score_model(clf):
    print ("\n분석모델: {}".format(clf.__class__.__name__))
    start = time.time()
    # use 3-fold CV
    scores = cross_validation.cross_val_score(clf, x_train, y_train, scoring='roc_auc', cv=3) 
    end = time.time()
    print ("소요시간 (secs): {:.3f}".format(end - start))
    print ("roc_auc: {:.3f}".format(scores.mean()))
    return scores.mean()
```

```python
# 분석모델 비교
scores = {}
# naive bayes
scores['naive_bayes'] = score_model(naive_bayes.GaussianNB())
# logistic regression
scores['logistic_regression'] = score_model(linear_model.LogisticRegression()) 

# ensemble methors
# Random Forest
scores['random_forest'] = score_model(ensemble.RandomForestClassifier()) 
# gradient boosting
scores['gradient_boosting'] = score_model(ensemble.GradientBoostingClassifier())
```

분석 성능이 가장 좋을 것으로 예상이 되는 GradientBoostingClassifier를 분석 모델로 사용

### 3.2.1 분석에 필요한 요소들 찾기(최대 feature 수)

```python
# 변수의 개수 찾기
print ("다차원 독립 변수 중 선택할 차원의 수는?") # 28개

# 최적값 찾기
max_feat = [10,20,30,40]
test_score = np.zeros(len(max_feat))

# 모델 만들기
for i, j in enumerate(max_feat):
    start = time.time()
    print ("개수: {:.3f}".format(j))
    # Setup a classifer
    clf = ensemble.GradientBoostingClassifier(max_features = j, 
                                              learning_rate=0.1,
                                              random_state=42)                 
    
    # k-fold cross-validationuse 사용( 4-fold CV)
    scores = cross_validation.cross_val_score(clf, x_train2, y_train2, scoring='roc_auc', cv=4)
    test_score[i] = scores.mean() # 평균값      
    end = time.time()
    print ("소요시간 (secs): {:.3f}".format(end - start))    
    print ("roc_auc: {:.5f}".format(scores.mean()))
    print

# 그래프 그리기
plt.figure(figsize=(10,8))
plt.plot(max_feat, test_score, lw = 2, label = 'Testing score')
plt.legend()
plt.title('max_features', fontsize=18, y=1.03)
plt.xlabel('max_features')
plt.ylabel('auc score')
```

### 3.2.2 분석에 필요한 요소 찾기(n_estimator)

```python
# n_estimators 설정
print ("n_estimators 설정") # 140개

# 최적값 찾기
n_estimators = [50,100,150,200]
test_score = np.zeros(len(n_estimators))

# 모형 만들기
for i, s in enumerate(n_estimators):
    start = time.time()
    print ("n_estimators의 개수: {:.3f}".format(s))
    # setup a classifer
    clf = ensemble.GradientBoostingClassifier(n_estimators = s, 
                                              learning_rate= 0.1,
                                              max_features = 28,
                                              random_state=42)                 
    # k-fold cross-validationuse 사용( 4-fold CV)
    scores = cross_validation.cross_val_score(clf, x_train2, y_train2, scoring='roc_auc', cv=4)
    test_score[i] = scores.mean()  # report the mean       
    end = time.time()
    print ("소요시간 (secs): {:.3f}".format(end - start))    
    print ("roc_auc: {:.5f}".format(scores.mean()))
    print

# 그래프 그리기
plt.figure(figsize=(10,8))
plt.plot(n_estimators, test_score, lw = 2, label = 'Testing score')
plt.legend()
plt.title('n_estimators', fontsize=18, y=1.03)
plt.xlabel('n_estimators')
plt.ylabel('auc score')
```

### 3.2.3 분석에 필요한 요소 찾기(tree parameters; max_depth, min_samples_leaf)

```python
# Tree Parameters 찾기(max_depth, min_samples_leaf)
print ("tree parameters: max_depth,min_samples_leaf") # 3,4

# 값 찾기
min_samples_leaf= [1,3,4,5,6]
max_depth = [2,3,4,5]

# inicializing
test_score = np.zeros([len(max_depth),len(min_samples_leaf)]) 

# 모델 만들기
for i, s in enumerate(max_depth):
    for j, t in enumerate(min_samples_leaf):
        start = time.time()
        print ("max_depth: {:.3f}".format(s))
        print ("min_samples_leaf: {:.3f}".format(t))
        # Setup a classifer
        clf = ensemble.GradientBoostingClassifier(n_estimators = 140,
                                                  max_features = 28,
                                                  learning_rate=0.1,
                                                  min_samples_leaf = t, 
                                                  min_samples_split=2*t,
                                                  max_depth = s,
                                                  random_state=42)        
        # k-fold cross-validationuse 사용( 4-fold CV)
        scores = cross_validation.cross_val_score(clf, x_train2, y_train2, 
                                                  scoring='roc_auc', cv=4)
        test_score[i][j] = scores.mean() # report the mean          
        end = time.time()
        print ("소요시간 (secs): {:.3f}".format(end - start))    
        print ("roc_auc: {:.5f}".format(scores.mean()))
        print

# 그래프 그리기
plt.figure(figsize=(10,8))
for i in range (len(max_depth)):
    plt.plot(min_samples_leaf, test_score[i], lw = 2, 
             label = 'max_depth='+str(i+2))
plt.legend()
plt.title('tree parameters', fontsize=18, y=1.03)
plt.xlabel('min_samples_leaf')
plt.ylabel('auc score')
```

### 3.2.4 분석에 필요한 요소 찾기(subsample)

```python
# Subsample 설정
print ("Tuning subsample")

# 최적값 찾기
subsample = [0.7,0.8,0.9,0.95,0.98]
test_score = np.zeros(len(subsample))

# 모델 만들기
for i, s in enumerate(subsample):
    start = time.time()
    print ("subsample: {:.3f}".format(s))
    # Setup a classifer
    clf = ensemble.GradientBoostingClassifier(subsample = s,
                                              n_estimators = 140,
                                              max_features = 28,
                                              max_depth = 3,
                                              min_samples_leaf = 5, 
                                              min_samples_split=10,
                                              learning_rate=0.1,
                                              random_state=42)                 

    # k-fold cross-validationuse 사용( 4-fold CV)
    scores = cross_validation.cross_val_score(clf, x_train2, y_train2,
                                              scoring='roc_auc', cv=4)
    test_score[i] = scores.mean() # report the mean          
    end = time.time()
    print ("Runinng time (secs): {:.3f}".format(end - start))    
    print ("roc_auc: {:.5f}".format(scores.mean()))
    print

# 그래프 그리기
plt.figure(figsize=(10,8))
plt.plot(subsample, test_score, lw = 2, label = 'Testing score')
plt.legend()
plt.title('subsample', fontsize=18, y=1.03)
plt.xlabel('subsample')
plt.ylabel('auc score')
```

### 3.2.5 분석에 필요한 요소 찾기(learning rate)

```python
# fine tune the learning rate
print ("Fine tuning learning rate")

# 최적값 찾기
learn_rate = [0.1, 0.05, 0.025, 0.0125, 0.00625, 0.003125]
test_score = np.zeros(len(learn_rate))

# 모델 만들기
for i, s in enumerate(learn_rate):
    start = time.time()
    print ("learn_rate: {:.3f}".format(s))
    # Setup a classifer
    # increase n_estimators 
    clf = ensemble.GradientBoostingClassifier(learning_rate=s,
                                              n_estimators = 140*(i+1),
                                              max_features = 28, 
                                              max_depth= 3, 
                                              subsample = 0.9,
                                              min_samples_leaf=5,
                                              min_samples_split=10,
                                             random_state = 42)                
    
    # k-fold cross-validationuse 사용(4-fold CV)
    scores = cross_validation.cross_val_score(clf, x_train2, y_train2, 
                                              scoring='roc_auc', cv=4)
    test_score[i] = scores.mean() # report the mean        
    end = time.time()
    print ("Runinng time (secs): {:.3f}".format(end - start))    
    print ("roc_auc: {:.5f}".format(scores.mean()))
    print

# 그래프 그리기
plt.figure(figsize=(10,8))
plt.plot(learn_rate, test_score, lw = 2, label = 'Testing set')
plt.legend()
plt.title('Learning rate', fontsize=18, y=1.03)
plt.xlabel('Learning rate')
plt.ylabel('auc score')
```

## 3.3 Machine Learing

```python
from sklearn import metrics
from sklearn.metrics import roc_curve, auc
```

```python
# choose performance metric
def performance_metric(y_true, y_pred):
    """ 실제값과 예측값 사이의 총 점수를 계산한 후 반환 """
    fpr, tpr, thresholds = metrics.roc_curve(y_true, y_pred[:,1])
    score = metrics.auc(fpr, tpr)
    return score
```

```python
# learning rate
def learning_curves(x, y, clf):
    """ plot the learning curve for the train and testing data set. """
    print ("Creating learning curve")
    
    # Create the figure 
    plt.figure(figsize=(10,8))
    
    # 훈련 세트를 다양하게 함
    sizes = [1000, 5000, 10000, 20000, 30000, 40000, 50000, 60000, 76020]
    train_score = np.zeros(len(sizes)) # train set의 점수
    test_score = np.zeros(len(sizes)) # test set의 점수
    
    # 모델 만들기
    for i, s in enumerate(sizes):
            print(s)    
            
            # train set의 ferformance 찾기(4-fold CV 이용)
            scores = cross_validation.cross_val_score(clf, x[:s], y[:s], scoring='roc_auc', cv=4)
            test_score[i] = scores.mean()    
            print ("train roc_auc: {:.5f}".format(test_score[i])) 

            # train set의 ferformance 찾기(모든 training data 이용)
            clf.fit(x[:s], y[:s])
            y_pred = clf.predict_proba(x[:s])
            train_score[i] = performance_metric(y[:s], y_pred)
            print ("test roc_auc: {:.5f}".format(train_score[i]))
            print

    # 그래프 그리기
    plt.figure(figsize=(10,8))
    plt.plot(sizes, train_score, lw = 2, label = 'Training set')
    plt.plot(sizes, test_score, lw = 2, label = 'Testing set')
    plt.legend()
    plt.title('Learning Curve', fontsize=18, y=1.03)
    plt.xlabel('number of samples')
    plt.ylabel('auc score')
```

```python
clf = ensemble.GradientBoostingClassifier(learning_rate=0.05,
                                          n_estimators = 140,
                                          max_features = 28,
                                          subsample = 0.9,
                                          max_depth= 3,                                           
                                          min_samples_leaf=5,
                                          min_samples_split=10,
                                         random_state = 42)
learning_curves(x_train2, y_train2,clf)
```
